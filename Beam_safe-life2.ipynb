# Wing's main beam calculation - Safe-life design


```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline

# Initial properties

a = 1
b = 1

MTOW = 94000
WingSpan = 36


# Material properties
StressYield = 290e6
FatigueStrenght = 140e06

# Thickness
th = np.arange(5e-3, 0.5, 5e-3)
```

## Loading

```python
Moment = (1/3)*(MTOW*9.81/2)*(WingSpan/2)
```

## Fixed-end beam with distributed force

```python
# Inertia

InertiaEx = (a*b**3)/12
InertiaIn = ((a - 2*th)*(b - 2*th)**3)/12
Inertia = InertiaEx - InertiaIn
```

```python
StressActual = Moment*(b/2)/Inertia
SFLife = FatigueStrenght/StressActual
SFDisposal = StressYield/StressActual
```

## One-time use

```python
# Ploting
# 35 mm, aprox

plt.figure(figsize=(10,5))
plt.plot(th*1000, SFDisposal)
plt.xlabel('thickness [mm]')
plt.ylabel('Safety factor')
plt.xlim(0, 20)
plt.ylim(0, 2.5)
plt.grid()
plt.show()
```

## Safe-life design

```python
plt.figure(figsize=(10,5))
plt.plot(th*1000, SFLife)
plt.xlabel('thickness [mm]')
plt.ylabel('Safety factor')
plt.ylim(0, 2)
plt.xlim(0, 150)
plt.grid()
plt.show()

```

```python
plt.figure(figsize=(10,5))
plt.plot(th*1000, StressActual)
plt.xlabel('thickness [mm]')
plt.ylabel('Stress')
plt.ylim(0, 300e6)
plt.xlim(0, 500)
plt.grid()
plt.show()
```
